1. What is nf-core?

nf-core is a community effort to develop and maintain a curated set of analysis pipelines built using Nextflow.

2. How many pipelines are there currently in nf-core?

They are 139 pipelines that are currently available as part of nf-core. (nf-core website)

3. Are there any non-bioinformatic pipelines in nf-core?

There are also other piplines in nf-core, like the meerpipe which is a astronomy pipeline that processes MeerKAT pulsar data to produce images and data products for pulsar timing analysis.

4. Let's go back a couple of steps. What is a pipeline and what do we use it for?

A bioinformatics pipeline is a set of connected algorithms (or blocks) that are executed in a predefined order, It is used for example to process and analyze next-generation sequencing (NGS) data locally or in cluster environments.

5. Why do you think nf-core adheres to strict guidelines?

They adhere to strict guidlines to ensure that the pipeline works stable, always and anywhere. 

6. What are the main features of nf-core pipelines?

nf-core provides a standardized set of best practices, guidelines, and templates for building and sharing scientific pipelines.
